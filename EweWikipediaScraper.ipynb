```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Because a list of categories is not available, we directly scrape using the pages listing the articles
url1 = "https://ee.wikipedia.org/w/index.php?title=Special:AllPages&from=ATI" #Starting url
url2 = "https://ee.wikipedia.org/w/index.php?title=Special:AllPages&from=Ronald+Reagan"
linksList = [url1, url2]
titlesList = []
```

### Scrape all articles in special index

```python
for currentLink in linksList:
    #Send request
    response = requests.get(currentLink)
    html = response.text
    pageSoup = BeautifulSoup(html,"html5lib")
    #Get list of article titles
    articles = pageSoup.find('ul', {'class' : 'mw-allpages-chunk'}).findAll('a')
    for article in articles:
        titlesList.append(article.text)
```
